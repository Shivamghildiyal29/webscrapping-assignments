```python
name =[]
year = []
genre =[]
run = []
rate =[]
vote = []

for i in driver.find_elements_by_xpath('//h3[@class="lister-item-header"]/a[1]'):
    name.append(i.text)

for i in driver.find_elements_by_xpath('//h3[@class="lister-item-header"]/span[2]'):
    year.append(i.text)

for i in driver.find_elements_by_xpath('//span[@class="genre"]'):
    genre.append(i.text)
    
for i in driver.find_elements_by_xpath('//span[@class="runtime"]'):
    run.append(i.text)

for i in driver.find_elements_by_xpath('//div[@class="ipl-rating-star small"]/span[2]'):
    rate.append(i.text)

for i in driver.find_elements_by_xpath('//p[@class="text-muted text-small"]/span[2]'):
    if  i.text != '|':
        vote.append(i.text)    
```
