### Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
url ='https://www.naukri.com/browse-jobs'
driver.get(url)

skill= driver.find_element_by_xpath('//*[@id="skill"]/div[1]/div[2]/input')
location =driver.find_element_by_xpath('//*[@id="location"]/div[1]/div[2]/input')

skill.send_keys('Data Analyst')
location.send_keys('Bangalore')

skill.submit()

title = []
location = []
company = []
experience = []

for i in driver.find_elements_by_xpath('//div[@class="info fleft"]/a[1]'):
    title.append(i.text)
    
for j in driver.find_elements_by_xpath('//ul[@class="mt-7"]/li[3]'):
    location.append(j.text)

for k in driver.find_elements_by_xpath('//div[@class="mt-7 companyInfo subheading lh16"]/a[1]'):
    company.append(k.text)

for l in driver.find_elements_by_xpath('//ul[@class="mt-7"]/li[1]'):
    experience.append(l.text)

title = title[0:10]
location = location[0:10]
company = company[0:10]
experience = experience[0:10]

q1 = pd.DataFrame(list(zip(title, location, company,experience)), 
               columns =["Title", 'Location', 'Company' , 'Experience'])

q1.head(10)
```

## 6 Scrape data of first 100 sunglasses listings on flipkart.com, attributes need to scrape are as follows:
1. Brand
2. Product Description
3. Price
4. Discount

```python
url = 'https://www.flipkart.com/'
driver.get(url)
```

```python
#closing the login prompt
driver.find_element_by_xpath('/html/body/div[2]/div/div/button').click()
```

```python
search = driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
search.send_keys('sunglasses')
```

```python
driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/button').click()
```

```python
brand =[]
product_desc = []
price =[]
org_price = []
discount = []
i=0
while i != 3:
    for j in driver.find_elements_by_xpath('//div[@class="_2B099V"]/div[1]'):
        brand.append(j.text)
        
    for k in driver.find_elements_by_xpath('//a[@class="IRpwTa"]'):    
        product_desc.append(k.text)
    
    for l in driver.find_elements_by_xpath('//div[@class="_25b18c"]/div[1]'):
        price.append(l.text)        

    for m in driver.find_elements_by_xpath('//div[@class="_25b18c"]/div[2]'):
        org_price.append(m.text)  
        
    for n in driver.find_elements_by_xpath('//div[@class="_25b18c"]/div[3]'):
        discount.append(n.text)  

    i = i+1   
    driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]').click()
    time.sleep(10)
```

```python
q6 = pd.DataFrame(list(zip(brand, product_desc, price, org_price, discount)), 
               columns =['Brand', 'Product Description', 'Price', 'Original price','Discount'])
    
```

## 9. scrape First 100 shoes data we get in the link- https://www.myntra.com/shoes. The data should include “Brand” of the shoes , Short Shoe description, price of the shoe

```python
url = '''https://www.myntra.com/shoes'''
driver.get(url)
```

```python
brand = []
desc = []
price = []

n = 0
while n!=2:
    for i in driver.find_elements_by_xpath('//h3[@class="product-brand"]'):
        brand.append(i.text)

    for j in driver.find_elements_by_xpath('//h4[@class="product-product"]'):
        desc.append(j.text)

    for k in driver.find_elements_by_xpath('//div[@class="product-price"]/span[1]'):
        price.append('Rs' + k.text.split('Rs')[1])
    n=n+1
    driver.find_element_by_xpath('//li[@class="pagination-next"]').click()
    time.sleep(10)
```

```python
q9 = pd.DataFrame(list(zip(brand, desc, price)), 
               columns =['Brand', 'Product Description', 'Price'])
q9    
```

## 10.Scraping laptops with filter “Intel Core i7” and “Intel Core i9” from link-https://www.amazon.in/. After setting the filters scrape first 10 laptops data. We have to scrape 3 attributes for each laptop:
* 1. title
* 2. Ratings
* 3. Price


```python
driver.get('https://www.amazon.in/')
search= driver.find_element_by_id("twotabsearchtextbox")
search.send_keys('laptop')
driver.find_element_by_xpath('//*[@id="nav-search-submit-button"]').click()
```

```python
driver.find_element_by_xpath('//*[@id="p_n_feature_thirteen_browse-bin/12598163031"]/span/a/span').click()
```

```python
links= []
for j in driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]'):
    links.append(j.get_attribute('href'))
```

```python
title = []
rating = []
price = []

for link in links:
    driver.get(link)
    
    title.append(driver.find_element_by_xpath('//*[@id="productTitle"]').text)

    try: 
        price.append(driver.find_element_by_xpath('//*[@id="priceblock_ourprice"]').text)
    except NoSuchElementException:
        price.append('Price Not available')
        
    try:
        rating.append(driver.find_element_by_xpath('//*[@id="acrPopover"]').get_attribute('title'))
    except NoSuchElementException:
        rating.append('No available rating')
    time.sleep(3)
```

```python
q10 = pd.DataFrame(list(zip(title, rating, price)), 
               columns =['Name', 'Rating', 'Price'])
q10    
```
